```python
#Tokenization based on space
data_list1 = list()
for row in df['comment']:
    data_list1.append(RegexpTokenizer('\w+').tokenize(row))
```

```python
#Tokenization based on space
data_list2 = list()
for row in df['parent_comment']:
    data_list2.append(RegexpTokenizer('\w+').tokenize(row))
```

```python
#converting to lower case
data_list_lower1 = []
for i in data_list1:
    lower = list(map(lambda x:x.lower(),i))
    data_list_lower1.append(lower)   
```

```python
#converting to lower case
data_list_lower2 = []
for i in data_list2:
    lower = list(map(lambda x:x.lower(),i))
    data_list_lower2.append(lower)   
```

```python
#remove puncts
from string import punctuation
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.stem import LancasterStemmer

nltk.download('omw-1.4')
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = list(stopwords.words('english'))
```

```python
punctuation
```

```python
final_stop_words = list(punctuation) + stop_words
```

```python
data_list_lower_stpwords1 = []
for line in data_list_lower1:
    stpwords = []
    for w in line:
        if w not in final_stop_words:
            stpwords.append(w)
    data_list_lower_stpwords1.append(stpwords)  
```

```python
data_list_lower_stpwords2 = []
for line in data_list_lower2:
    stpwords = []
    for w in line:
        if w not in final_stop_words:
            stpwords.append(w)
    data_list_lower_stpwords2.append(stpwords)  
```

```python
data_list_lower_stpwords1[0:2]
```

```python
data_list_lower_stpwords2[0:2]
```

```python
#Lemmatization
lemma = WordNetLemmatizer()
data_list_lower_stpwords_lem1 = []
for line in data_list_lower_stpwords1:
    lines = list(map(lambda x:lemma.lemmatize(x),line))
    data_list_lower_stpwords_lem1.append(lines)
```

```python
data_list_lower_stpwords_lem1[0:2]
```

```python
#Lemmatization
lemma = WordNetLemmatizer()
data_list_lower_stpwords_lem2 = []
for line in data_list_lower_stpwords2:
    lines = list(map(lambda x:lemma.lemmatize(x),line))
    data_list_lower_stpwords_lem2.append(lines)
```

```python
data_list_lower_stpwords_lem2[0:2]
```

```python
df['CleanCommentToken'] = data_list_lower_stpwords_lem1
```

```python
df['CleanParentCommentToken'] = data_list_lower_stpwords_lem2
```

```python
sent = []
for line in data_list_lower_stpwords_lem1:
    seq = ''
    for word in line:
        seq = seq + ' ' + word
    sent.append(seq)
final_comment = sent
```

```python
sent = []
for line in data_list_lower_stpwords_lem2:
    seq = ''
    for word in line:
        seq = seq + ' ' + word
    sent.append(seq)
final_parent_comment = sent
```

```python
df['CleanComment'] = final_comment
```

```python
df['CleanParentComment'] = final_parent_comment
```

```python
df[['CleanComment','CleanParentComment']]
```

```python
vectorizer1 = CountVectorizer().fit_transform(df['CleanComment'])
```

```python
vectorizer2 = CountVectorizer().fit_transform(df['CleanParentComment'])
```

```python
from sklearn.preprocessing import LabelEncoder
lb_make = LabelEncoder()
df['topic_encoded']=lb_make.fit_transform(df['topic'])
```

```python
df['top']=data['top']
```

```python
df['down']=data['down']
```

```python
df['score']=data['score']
```

```python
df.columns
```

```python
import scipy
X = scipy.sparse.hstack((vectorizer1, 
                         vectorizer2,
                         df[['top', 'down','score','topic_encoded']].to_numpy())).tocsr()
```

```python
X[0:2]
```

```python
y = df['label']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=111)
```

```python
X_train.shape,X_test.shape
```

```python
##### Create Model Model ######
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, recall_score, classification_report, cohen_kappa_score
from sklearn import metrics 
```

```python
# Baseline Random forest based Model
rfc = RandomForestClassifier(criterion = 'gini', n_estimators=1000, verbose=1, n_jobs = -1, 
                             class_weight = 'balanced', max_features = 'auto')
rfcg = rfc.fit(X_train,y_train) # fit on training data


####### Prediction ##########
predictions = rfcg.predict(X_test)
print('Baseline: Accuracy: ', round(accuracy_score(y_test, predictions)*100, 2))
print('\n Classification Report:\n', classification_report(y_test,predictions))
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver='lbfgs',max_iter=10000)
```
